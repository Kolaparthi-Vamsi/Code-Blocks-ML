PEICE OF CODES FOR COUNT VECTORIZERS AND TFIDF 

```python
#COUNT VECTORIZER 
#vectorizer converts the doc in the rows and columns. Rows --> number of docs or lines(=5) and cols--> all the unique words
docs=["the house had a tiny little mouse", 
"the cat saw the mouse,cat is intelligent", 
"the mouse ran away from the house, mouse is fast", 
"the cat finally ate the mouse,cat is very happy", 
"the end of the mouse story as mouse died"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#Parameters 
#      input--> corpus
#      stop words
#      n_gram range(unary, binary, etc) 
#      max_df, min_df
#      max_features
cv=CountVectorizer(stop_words=['the','of','a','as','is'])
X=cv.fit_transform(docs)
```

```python
cv.get_feature_names()# see the unique words and the stop words are not incluce here
```

```python
print(X)# this is the sparse matrix, this will store the non zero frequencies with respective to the positons it exists. Other zero frequency ignored
```

```python
X.shape# this will give us the array shape but it is stored in above format to preserve storage 
```

```python
X.toarray()# to convert to an array
```

```python
# FORM DATAFRAME
import pandas as pd
df=pd.DataFrame(X.toarray(),columns=cv.get_feature_names())
df.head()
```

```python
cv.vocabulary_
```

WHAT IS IDF? 
        IDF--> inverse document frequency
        -->log(number of documents or records/ number of records containing the word

```python
# IDF is important because it slashes out most used common words.But the importance of log is doesnot make the value diminished as it can hold value
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words=['the','of','a','as','is'])
Y=tf.fit_transform(docs)
Y.shape
```

```python
tf.vocabulary_
```

```python
tf.idf_# this gives the tfidf ie--> idf and these ar take only for non -zero elements
```

```python
Y.toarray()
```

```python
df1=pd.DataFrame(Y.toarray(),columns=cv.get_feature_names())
df1.head(4)
```

TFIDF TRANSFORMER ?


```python
from sklearn.feature_extraction.text import TfidfTransformer

tt=TfidfTransformer()
z=tt.fit(cv.fit_transform(docs))
z.idf_
```

```python
k=tt.transform(cv.fit_transform(docs))
k
```
